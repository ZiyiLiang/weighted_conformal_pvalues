```python
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [5, 5]

import numpy as np
import pandas as pd
import seaborn as sns
import statistics as stats
from tqdm import tqdm
import pdb
import copy
from scipy.io import arff, loadmat

import sys
import os
```

```python
## Load the credit card data
base_path = "/media/msesia/Samsung/data/"
data_raw = pd.pandas.read_csv(base_path + "creditcard_raw.csv", sep=",")
```

```python
data_raw
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X = np.array(data_raw.iloc[:,:-1])
Y = np.array(data_raw.iloc[:,-1])

scaler = StandardScaler()
X_scale = scaler.fit_transform(X)

pca2 = PCA(n_components=2)
principalComponents = pca2.fit_transform(X_scale)

df = pd.DataFrame(data = principalComponents, columns = ['PC 1', 'PC 2'])
df["Y"] = Y
df
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df['PC 1'],df['PC 2'],c=1-df['Y'],cmap='prism')
plt.xlabel('pc1')
plt.ylabel('pc2')
plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize=(7,3))
fig.tight_layout()

axs[0].hist(data_raw["Amount"][Y==0])
axs[0].set_yscale('log')
axs[1].hist(data_raw["Amount"][Y==1])
axs[1].set_yscale('log')

plt.show()
```

```python
data = copy.deepcopy(data_raw)
idx_in = np.where(data_raw["Class"]==0)[0]
data.iloc[idx_in,-1] = "normal"
idx_in = np.where(data_raw["Class"]==1)[0]
data.iloc[idx_in,-1] = "fraud-0"
idx_out_g1 = np.where((data_raw["Amount"] > 100)*(data_raw["Class"]==1)==1)[0]
data.iloc[idx_out_g1,-1] = "fraud-1"
```
